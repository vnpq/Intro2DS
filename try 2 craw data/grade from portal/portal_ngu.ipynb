```python
# Phân tích HTML bằng BeautifulSoup
soup = BeautifulSoup(page_html, "html.parser")

# Tìm bảng theo class
table = soup.find('table', {'id': 'tbDiemThiGK'})

# Lấy tiêu đề cột từ thead
headers = [th.text.strip() for th in table.find("thead").find_all("th")]

# Lấy dữ liệu từ tbody
rows = []
for row in table.find("tbody").find_all("tr"):
    cols = [col.text.strip() for col in row.find_all("td")]
    rows.append(cols)

# Tạo DataFrame từ dữ liệu đã trích xuất
df = pd.DataFrame(rows, columns=headers)

df
```

```python
dftypes = df.dtypes
dftypes
```

```python
df['Số TC'] = df['Số TC'].astype(int)
df['Điểm'] = df['Điểm'].str.replace(",", ".")
df['Điểm'] = pd.to_numeric(df['Điểm'], errors="coerce")

split_columns = df['Môn Học'].str.split(" - ", expand=True)
df['Mã môn'] = split_columns[0]
df['Tên môn'] = split_columns[1]
df = df.drop(columns=["Môn Học", "Mã LĐ", "Lớp", "Ghi Chú"])
df = df[['NH/HK', 'Mã môn', 'Tên môn', 'Số TC', 'Điểm']]
df = df.drop(index=9)
```

```python
df.to_csv("portal_ngu.csv", sep=',', encoding='utf-8', index=False, header=True)
df
```

```python
# tính gpa nhưng không tính các môn có code: BAA00021. BAA00022, BAA00030
not_include = ["BAA00021", "BAA00022", "BAA00030"]
gpa = 0
total_credit = 0
for index, row in df.iterrows():
    if row['Mã môn'] in not_include:
        continue
    total_credit += row["Số TC"]
    gpa += row["Điểm"] * row["Số TC"]
gpa = gpa / total_credit

print("GPA: ", gpa)
```
