```python
# Phân tích HTML bằng BeautifulSoup
soup = BeautifulSoup(page_html, "html.parser")

# Tìm bảng theo class
table = soup.find("table", {"class": "table table-hover table-bordered table-striped"})

# Lấy tiêu đề cột từ thead
headers = [th.text.strip() for th in table.find("thead").find_all("th")]

# Lấy dữ liệu từ tbody
rows = []
for row in table.find("tbody").find_all("tr"):
    cols = [col.text.strip() for col in row.find_all("td")]
    rows.append(cols)

# Tạo DataFrame từ dữ liệu đã trích xuất
df = pd.DataFrame(rows, columns=headers)

df
```

```python
dftypes = df.dtypes
dftypes
```

```python
df = df.rename(columns={
    "10": "Total"
})
df = df.drop(columns=["Lớp", "Loại", "HL"])

df['Số TC'] = df['Số TC'].astype(int)
grade_cols = ['GK', 'TH', 'CK', 'Cộng', 'Khác', 'Total']
for feature in grade_cols:
    df[feature] = df[feature].str.replace(",", ".")
    df[feature] = pd.to_numeric(df[feature], errors="coerce")
    
dftypes = df.dtypes
dftypes
```

```python
df.to_csv("portal_ctdb.csv", sep=',', encoding='utf-8', index=False, header=True)
df
```

```python
# tính gpa nhưng không tính các môn có code: BAA00021. BAA00022
gpa = 0
total_credit = 0
for index, row in df.iterrows():
    if row['Mã MH'] == "BAA00021" or row['Mã MH'] == "BAA00022":
        continue
    total_credit += row["Số TC"]
    gpa += row["Total"] * row["Số TC"]
gpa = gpa / total_credit

print("GPA: ", gpa)
```
